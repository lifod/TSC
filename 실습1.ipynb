```python
import csv
f=open('/content/data.csv',encoding='cp949')
data=csv.reader(f)
next(data)
date=[]
high=[]
low=[]

for row in data :
  if row[-1] !='':
    if row[0].split('-')[1]=='12' and row[0].split('-')[2]=='21':
      high.append(float(row[-1]))
      date.append(row[0].split('-')[0])

  if row[-2]!='':
    if row[0].split('-')[1]=='12' and row[0].split('-')[2]=='21':
      low.append(float(row[-2]))

print(high)
print(low)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(25,8))
plt.title('내 생일 기온 비교',size=20)
plt.plot(date,high,'red',label='high')
plt.plot(date,low,'blue',label='low')
plt.xticks(size=13)
plt.yticks(size=20)
plt.legend()
plt.show()
```
